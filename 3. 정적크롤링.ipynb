## 속성 데이터

```python
html = """<html> <head><title class="t" id="ti">test site</title></head> <body> <p>test</p> <p>test1</p> <p>test2</p> </body></html>"""
```

- html
    - head
        - title
    - body
        - p
        - p
        - p

```python
soup = BeautifulSoup(html, "lxml")
```

```python
soup
```

```python
# soup에서 처음으로 등장하는 title태그
tag_title = soup.title

print(tag_title)
print(tag_title.attrs) # 태그의 속성 가져오기
print(tag_title["class"])
print(tag_title["id"])
```

```python
# 해당 속성이 없다면 에러 발생
tag_title["class1"]
```

```python
# tag타입은 딕셔너리처럼 접근할 수 있고 딕셔너리 문법을 그대로 적용할 수 있음
tag_title.get("class1", "default_value")
```

### 태그 접근

- soup.태그이름 의 형태로 첫 번째로 등장하는 태그의 정보를 가져올 수 있음

```python
tag_title = soup.title # title 태그
print(tag_title)
```

```python
# 태그의 텍스트 값 출력
print(tag_title.text)
print(tag_title.string)
```

```python
# text와 string의 차이
html = """<html> <head><title>test site</title></head> <body> <p><span>test1</span><span>test2</span></p> </body></html>"""
```

- html
    - head
        - title
    - body
        - p
            - span
            - span

```python
soup = BeautifulSoup(html, "lxml")

tag_p = soup.p

data_text = tag_p.text
data_string = tag_p.string

print("text: ", data_text, type(data_text))
print("string: ", data_string, type(data_string))
```

- text
    - 하위 태그들의 값도 모두 출력
    
- string
    - 정확히 해당 태그에 대한 값만 출력

```python
tag_p.span.string # p태그 안에 있는 첫 번째 span태그의 텍스트 값
```

## 원하는 요소에 접근하기

### find_all()

- 원하는 태그들을 리스트 형태로 가져오기

```python
html = """<html> <head><title>test site</title></head> <body> <p id="i" class="a">test1</p><p id="d" class="d">test2</p><p class="c">test3</p></p><a>a tag</a> <b>b tag</b></body></html>"""
```

- html
    - head
        - title
    - body
        - p
        - p
        - p
        - a
        - b

```python
soup = BeautifulSoup(html, "lxml")
```

```python
# find_all로 title태그 찾기
soup.find_all("title")
```

```python
# find_all로 p태그 찾기
soup.find_all("p")
```

- id값으로 태그 가져오기

```python
# id가 "d" 인 태그 가져오기
soup.find_all(id = "d")
```

```python
# id의 존재 여부로 데이터 가져오기
soup.find_all(id = True)
```

- 원하는 태그, 원하는 id값으로 태그 가져오기

```python
print(soup.find_all("p", id = "d"))
print(soup.find_all("p", id = "c"))
```

```python
# 원하는 태그, 원하는 class값으로 태그 가져오기
print(soup.find_all("p", class_ = "d"))
print(soup.find_all("p", class_ = "c"))
```

- find_all() 연속으로 사용하기

```python
# body 태그 안에 있는 p태그만 찾고 싶은 경우
tag_body = soup.find_all("body")
print(tag_body)
tag_p = tag_body[0].find_all("p")
print(tag_p)
```

### find()

- 하나의 요소만 가져옴
- 찾고자 하는 요소가 하나만 있을 때 주로 사용
    - 예) id값으로 접근

```python
soup.find("p")
```

```python
print(soup.find("p", class_ = "d"))
print(soup.find("p", id = "i"))
```

```python
# 연속으로 find() 사용
soup.find("body").find("p", class_ = "d")
```

### select()

- find_all()과 마찬가지로 매칭되는 모든 결과를 리스트로 반환
- select_one()으로 하나의 결과만 반환하는 것도 가능
- 클래스는 마침표(.), 아이디는 샵(#), 자손태그는 띄어쓰기로 표현, 자식태그는 (>) 로 표현

```python
print(soup.select("p")) # p 태그들
print(soup.select(".d")) # 클래스가 d인 태그들
print(soup.select("p.d")) # 클래스가 d인 p태그들
print(soup.select("#i")) # 아이디가 i인 태그들
print(soup.select("p#i")) # 아이디가 i인 p태그들
```

```python
html = """<html> <head><title>test site</title></head> <body> <div><p id="i" class="a">test1</p><p class="d">test2</p></div><p class="d">test3</p> <a>a tag</a> <b>b tag</b></body></html>"""
```

- html
    - head
        - title
    - body
        - div
            - p
            - p
        - p
        - a
        - b

```python
soup = BeautifulSoup(html, "lxml")
```

```python
print(soup.select("body p")) # body의 자손인 p태그들
print(soup.select("body .d")) # body의 자손인 클래스가 d인 태그들
print(soup.select("body p.d")) # body의 자손 중에 클래스가 d인 p태그들
print(soup.select("body #i")) # body의 자손 중에 아이디가 i인 태그들
print(soup.select("body p#i")) # body의 자손 중에 아이디가 i인 p태그들
print(soup.select("div p")) # div의 자손인 p태그들
```

```python
print(soup.select("p[id]")) # id가 존재하는 p태그들
print(soup.select("p[class]")) # class가 존재하는 p태그들
```

## 예제1. 티스토리 크롤링

```python
# 제목 수집
soup.select_one("div.hgroup > h1").string
```

```python
# 게시물 내용 수집(Hello, world!)
soup.select_one("div.tt_article_useless_p_margin.contents_style > p").string
```
