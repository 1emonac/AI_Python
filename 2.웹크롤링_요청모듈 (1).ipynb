```python
import requests
```

# requests 로 웹페이지 접속

```python
url = "https://www.naver.com"

requests.get(url)
```

```python
url = "https://www.naver.com/a"

res = requests.get(url)
```

```python
# 응답코드 확인하기
print(res.status_code)
```

```python
# 헤더 가져오기
url = "https://www.naver.com"

res = requests.get(url)

print(res)
print(res.headers)
```

```python
# html코드 보기
print(res.text)
```

# urllib으로 웹페이지 접속

```python
from urllib.request import urlopen
```

```python
url = "https://www.naver.com"

page = urlopen(url)

print(page)
```

```python
print(page.code)
print(page.headers)
print(page.url)
print(page.info().get_content_charset())
```

```python
print(page.read()) # html코드 확인
```
