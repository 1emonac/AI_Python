# **파이썬으로 CSV 파일 출력하기**

```python
with open('남산도서관 장서 대출목록 (2021년 04월).csv') as f:
  print(f.readline())
```

UnicodeDecodeError: 'utf-8' codec can't decode byte 0xb9 in position 0: invalid start byte

파이썬의 open() 함수는 기본적으로 텍스트 파일이 UTF-8 형식으로 저장되어 있다고 가정합니다.
하지만 한글 텍스트는 여전히 완성형 인코딩인 EUC-KR을 사용하는 일이 잦아 오류가 발생.

# **파일 인코딩 형식 확인하기: chardet.detect()함수**

```python
import chardet
with open('남산도서관 장서 대출목록 (2021년 04월).csv', mode='rb') as f:
    d = f.readline()
print(chardet.detect(d))
```

chardet 패키지의 chardet.detect()함수를 사용.
-> 문자 인코딩 방식을 알아낼 수 있음

mode 매개변수를 바이너리 읽기 모드인 'rb'로 지정.
-> open() 함수 mode 매개변수의 기본값은 텍스트 읽기 모드인 'r'

# **인코딩 형식 지정하기**

```python
with open('남산도서관 장서 대출목록 (2021년 04월).csv', encoding ='EUC-KR') as f:
    print(f.readline())
    print(f.readline())
```

open() 함수로 파일을 읽을 때 encoding 매개변수로 인코딩 형식을 'EUC-KR'로 지정

# **CSV 파일을 데이터프레임으로 읽기: read_csv() 함수**

```python
import pandas as pd
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR')
```

 DtypeWarning: Columns (5,6,9) have mixed types. Specify dtype option on import or set low_memory=False.

 자동으로 파악한 데이터 타입이 달라지면서 경고가 발생함.

 'ISBN', '세트 ISBN', '주제분류번호'에 해당하는 5,6,9 열 때문에 오류가 발생함.
 해결 방법은 **low_memory 매개변수**를 False로 지정하여 파일을 나누어 읽지 않고 한 번에 읽는 것.

```python
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR', low_memory=False)
```

low_memory 매개변수를 False로 지정하면 경고는 발생하지 않지만, CSV 파일을 한번에 모두 읽기 때문에 많은 메모리를 사용합니다. CSV 파일이 아주 큰 경우 메모리 부족 오류가 발생할 수 있습니다. low_memory 매개변수를 사용하지 않는 다른 방법은 열의 데이터 타입을 자동으로 찾지 않도록 아예 dtype 매개변수로 데이터 타입을 지정하는 것입니다. 경고가 발생했던 5, 6, 9 열에 해당하는 열의 데이터 타입을 문자열로 지정하여 읽는 코드는 다음과 같습니다.

df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='euc-kr', dtype={'ISBN': str, '세트 ISBN': str, '주제분류번호': str})

```python
df.head()
```

마지막 '**Unnamed: 13**'은 열 마지막에 콤마(,)가 붙었기 때문에 판다스가 이 콤마를 보고 마지막에 하나의 열이 더 있다고 판단하여
'Unnamed: 13'열이 표시됨

### **판다스의 read_table() 함수**

CSV 파일을 읽을 때는 **read_table() 함수도** 사용할 수 있음.
read_table() 함수는 read_csv() 함수와 매우 비슷합니다. 콤마(,) 대신 탭(\t)으로 구분된 파일을 읽는 데 사용합니다. 사실 read_table() 함수는 구분자로 탭을 사용하는 read_csv(sep='\t')와 같습니다.

# **데이터프레임을 CSV 파일로 저장하기: to_csv() 메서드**

```python
df.to_csv('ns_202104.csv')
```

판다스의 데이터프레임을 CSV로 저장할 때는 to_csv() 메서드를 사용합니다. to_scv() 메서드는 기본적으로 UTF-8 형식으로 저장하기 때문에 나중에 open() 함수로 파일의 내용을 읽을 때 따로 encoding 매개변수를 사용하지 않아도 됩니다.

```python
with open('ns_202104.csv') as f:
    for i in range(3):
        print(f.readline(), end='')
```

```python
ns_df = pd.read_csv('ns_202104.csv', low_memory=False)
ns_df.head()
```

이전에 생성되었던 인덱스가 'Unnamed: 0'으로 저장되어 표시됨

```python
ns_df = pd.read_csv('ns_202104.csv', index_col=0, low_memory=False)
ns_df.head()
```

index_col 매개변수를 사용하여 첫 번째 열에 인덱스가 있으므로 0으로 지정

또 다른 방법은 애초에 데이터프레임을 CSV 파일로 저장할 때 인덱스를 빼고 저장할 수 있습니다.
이렇게 하려면 다음처럼 to_csv() 메서드에 index 매개변수를 False로 지정합니다.

```python
df.to_csv('ns_202104.csv', index=False)
```
